```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("haberman.csv")
data.head()
```

```python
# Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(data["Age"], bins=50, kde=False)
plt.title("Age vs Survival Status")
plt.xlabel("Age")
plt.ylabel("Survival Status")
plt.show()
```

Observation:

The x-axis represents age, while the y-axis is labeled as "Survival Status," suggesting that the plot may intend to show how age distribution relates to survival status

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x="Year", palette="Set1")
plt.title("Year of Operation Distribution")
plt.xlabel("Year of Operation")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

Observation:
The figure shows a count plot of surgical operations performed in different years. It suggests that certain years had a notably higher frequency of operations, while others had fewer

```python
plt.figure(figsize=(10, 6))
sns.histplot(data["No.of Pos Axillary nodes"], bins=10, kde=True)
plt.title("Positive Axillary Nodes Distribution")
plt.xlabel("Number of Positive Nodes")
plt.ylabel("Frequency")
plt.show()
```

Observation:
The figure displays a right-skewed histogram of the "Number of Positive Axillary Nodes," indicating that most individuals have a low number of positive nodes. This distribution can be relevant in a medical context for understanding prognosis and treatment outcomes.

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=data, x="Survival Status")
plt.title("Survival Status Distribution")
plt.xlabel("Survival Status")
plt.ylabel("Count")
plt.xticks(ticks=[0, 1], labels=["Survived", "Not Survived"])
plt.show()
```

Observation:

It shows that there are more cases classified as "Survived" compared to "Not Survived," providing a basic overview of the dataset's survival outcomes.

```python
# Pairwise relationships
sns.pairplot(data, hue="Survival Status", diag_kind="kde")
plt.suptitle("Pairwise Relationships")
plt.show()
```

Observation:
The pairplot visually examines relationships between pairs of numeric variables in the dataset while color-coding points based on "Survival Status." 

```python
# Box plots
plt.figure(figsize=(6, 6))
sns.boxplot(data=data, x="Survival Status", y="Age")
plt.title("Age vs. Survival Status")
plt.xlabel("Survival Status")
plt.ylabel("Age")
plt.xticks(ticks=[0, 1], labels=["Survived", "Not Survived"])
plt.show()
```

Observation:
The box plot "Age vs. Survival Status" shows that the median age is similar for both groups (Survived and Not Survived). 

```python
# Scatter plot: Age vs. Positive Nodes
plt.figure(figsize=(12, 5))
sns.scatterplot(data=data, x="Age", y="No.of Pos Axillary nodes", hue="Survival Status")
plt.title("Age vs. Positive Axillary Nodes")
plt.xlabel("Age")
plt.ylabel("Number of Positive Nodes")
plt.legend(title="Survival Status", labels=["Survived", "Not Survived"])
plt.show()
```

Observation:
The scatter plot "Age vs. Positive Axillary Nodes" shows no strong linear correlation between age and the number of positive axillary nodes. For "Survived" cases, there's a concentration of fewer nodes across ages, while "Not Survived" cases vary more in node counts regardless of age. The number of positive nodes appears more influential in survival outcomes.


```python
# Correlation matrix
correlation_matrix = data.corr()
correlation_matrix
```

```python
# Heatmap of correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Violin plot: Survival Status vs. Age
plt.figure(figsize=(8, 6))
sns.violinplot(data=data, x="Survival Status", y="Age")
plt.title("Survival Status vs. Age")
plt.xlabel("Survival Status")
plt.ylabel("Age")
plt.xticks(ticks=[0, 1], labels=["Survived", "Not Survived"])
plt.show()
```

Observation:

Survivors (labeled as "Survived") generally exhibit a more concentrated age distribution, with a peak in the younger age range.
Non-survivors (labeled as "Not Survived") have a broader and flatter age distribution, indicating a wider spread of ages among this group.

```python
# Probability density function: Positive Nodes vs. Survival Status
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x="No.of Pos Axillary nodes", hue="Survival Status", element="step", stat="probability")
plt.title("Survival Probability by Positive Nodes")
plt.xlabel("Number of Positive Nodes")
plt.ylabel("Probability")
plt.legend(title="Survival Status", labels=["Survived", "Not Survived"])
plt.show()
```

Observation:

As the number of positive axillary nodes increases, the probability of not Survive  also increases significantly, while the probability of Survive  decreases.

```python
# Statistical Test: Mann-Whitney U Test

# The Mann-Whitney U test can be performed to determine if there are statistically significant differences between the age distributions of patients who survived and those who did not.
from scipy.stats import mannwhitneyu

survived_age = data[data["Survival Status"] == 1]["Age"]
not_survived_age = data[data["Survival Status"] == 2]["Age"]

statistic, p_value = mannwhitneyu(survived_age, not_survived_age)
print("Mann-Whitney U Test:")
print(f"Statistic: {statistic}")
print(f"P-value: {p_value}")
```

Observation: Mann-Whitney U Statistic: The calculated Mann-Whitney U statistic quantifies the difference between the age distributions of the two groups. However, it's important to note that the magnitude of the statistic itself does not indicate the direction or significance of the difference.
